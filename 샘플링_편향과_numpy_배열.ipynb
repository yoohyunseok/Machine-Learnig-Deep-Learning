```python
fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8,
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0] #생선 길이 리스트 생성
fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7,
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9] #생선 무게 리스트 생성
```

```python
fish_data = [[l, w] for l, w in zip(fish_length, fish_weight)] #생선 무게와 길이 데이터 2차원 리스트로 생성
#이때 하나의 생선 데이터를 샘플(sample) 이라 부른다. 총 49개의 sample이 있는 것.
#이 샘플은 무게와 길이, 두개의 feature을 사용함.
fish_target = [1]*35 + [0]*14 #생선 정답 리스트 생성
```

# KNeighborsClassifier 클래스 임포트 및 객체 생성

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
```

# index 사용해서 샘플 출력

```python
print(fish_data[4])
```

# 슬라이싱 기능을 이용해 샘플 출력

```python
print(fish_data[0:5])#0~4
print(fish_data[:5])#처음부터 시작되는거면 0 생략 가능
```

```python
print(fish_data[44:]) #같은 이치로 마지막 샘플까지 쭉 포함할 시 두번 째 인덱스 생략 가능
```

# 훈련, 테스트 세트 나누기

```python
train_input = fish_data[:35] #훈련 세트 인풋
train_target = fish_target[:35] #훈련 세트 타겟
test_input = fish_data[35:] #테스트 세트 인풋
test_target = fish_target[35:] #테스트 세트 타겟
```

```python
kn.fit(train_input, train_target) #훈련 세트로 학습
kn.score(test_input, test_target) #테스트 세트로 평가
```

# 샘플링 편향 발생

*   샘플링 편향: 훈련 세트와 테스트 세트에 샘플이 고루 섞여있지 않아서 샘플링이 한쪽으로 치우친 현상
*   이 예시에서는 훈련 세트에는 도미 샘플만, 테스트 세트에는 빙어 샘플만 들어가있음

# Numpy 사용해서 샘플링 편향 해결

```python
import numpy as np
```

```python
input_arr = np.array(fish_data) #numpy 배열로 변환
target_arr = np.array(fish_target)
```

```python
print(input_arr.shape) #배열의 크기를 알려주는 shpae attribute
print(input_arr) #(49, 2) => (행 개수, 열 개수) cardinality, degree?
```

```python
np.random.seed(42) #numpy의 무작위 결과를 만드는 함수들은 실행 될 때 마다 다른 결과를 만든다. 일정한 결과를 얻을려면 randomsed를 지정해야함
index = np.arange(49) #0~48 1씩 증가하는 인덱스 생성
np.random.shuffle(index) #인덱스 섞기
print(index)
```

**numpy의 배열 인덱싱 기능**

```python
print(input_arr[[1, 3]]) #한번에 여러 개의 인덱스로 여러개의 element를 선택 가능
```

**훈련 세트**

```python
train_input = input_arr[index[:35]]
train_target = target_arr[index[:35]]
```

```python
print(input_arr[13], train_input[0])
```

**테스트 세트**

```python
test_input = input_arr[index[35:]]
test_target = target_arr[index[35:]]
```

**샘플들이 잘 섞였는지 확인**

```python
import matplotlib.pyplot as plt
plt.scatter(train_input[:,0], train_input[:,1]) #2차원 배열은 행과 열 인덱스를 콤마(,)로 나누어 지정
plt.scatter(test_input[:,0], test_input[:,1])
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

**모델 훈련 및 검증**

```python
kn.fit(train_input, train_target)
```

```python
kn.score(test_input, test_target) #정확도
```

```python
kn.predict(test_input) #예측
```

```python
print(test_target) #테스트 세트에 대한 예측 결과가 일치함
```
